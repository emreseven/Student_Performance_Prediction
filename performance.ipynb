Import necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

read the dataset

```python
readdata = pd.read_csv("Student_Performance.csv")
data = readdata.copy()
data.head()
```

We readed our data successfully, now I will check for null values and outliers

```python
data.isnull().sum()
```

```python
data.describe()
```

no null values, descriptive numbers looks okey

I will encode extracurricular activities column 

```python
data["Extracurricular Activities"].value_counts()
```

```python
data["Extracurricular Activities"] = data["Extracurricular Activities"].map({"Yes": 1, "No": 0})
data.head(20)
```

Now it is time to split our dataset and train our model

```python
Y = data["Performance Index"]
X = data.drop(columns=["Performance Index"])
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
r2_score(y_test, y_pred)
```

I will dump my model

```python
joblib.dump(model,"model.pkl")
```
